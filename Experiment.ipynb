```python
import numpy as np
import cv2 as cv

import matplotlib.pyplot as plt
# %matplotlib inline  

def showImage(img, im_name) :
    fig = plt.figure()
    ax1 = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax1.set_title(im_name)
    plt.imshow(img, cmap='gray')
```

```python
face_cascade = cv.CascadeClassifier('face_dataset/haarcascade_frontalface_default.xml')
# raw_img = cv.imread('face_dataset/FacePhoto/pg42RF.jpg')
raw_img = cv.imread('face_dataset/hilmi.jpg')
img = cv.cvtColor(raw_img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(raw_img, cv.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
showImage(img, "Face")
```

```python
import skin_detector

skin = skin_detector.process(img)

showImage(skin, "skin")
```
